```python
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
import pandas as pd

tagged_data = [TaggedDocument(words=str(_d).split(), tags=[str(i)]) for i, _d in enumerate(dataset.token)]

max_epochs = 10
vec_size = 128
alpha = 0.025

model = Doc2Vec(size=vec_size,
                alpha=alpha, 
                min_alpha=0.00025,
                min_count=10,
                dm =1)

model.build_vocab(tagged_data)

for epoch in range(10):

    print('iteration {0}'.format(epoch))
    
    model.train(tagged_data,
                total_examples=model.corpus_count,
                epochs=model.iter)
    
    model.alpha -= 0.0002
    model.min_alpha = model.alpha

print("Model Trained")
```
